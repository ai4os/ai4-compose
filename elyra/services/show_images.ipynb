```python
import os
import base64
import matplotlib.pyplot as plt
import io
from PIL import Image
```

# Define path and name of the files

```python
input_file_path = os.path.join('..', 'output', 'text', os.environ.get('FILE_INPUT'))
output_file_path = os.path.join('..', 'output', 'images', os.environ.get('FILE_OUTPUT'))
```

# Read the file text

```python
with open(input_file_path, 'r') as file:
    lines = file.readlines()
```

# Obtain the last lane that contains the base64 coded image

```python
image_data = lines[-1]
```

# Decode image from base64

```python
image_bytes = base64.b64decode(image_data)
```

# Create a PIL image from bytes

```python
image = Image.open(io.BytesIO(image_bytes))
```

# Save image as a png

```python
image.save(output_file_path, 'png')
```

# Display image

```python
plt.imshow(image)
plt.show()
```
